```python
import cv2
import numpy as np

# Read the input image
input_image_path = '/Users/nikhilvalsankulangareth/Downloads/original_image_1.jpg'
img = cv2.imread(input_image_path, cv2.IMREAD_COLOR)

# Ensure the image is loaded successfully
if img is None:
    print("Error: Unable to read the input image.")
    exit()

# Define a function to apply and save a filter
def apply_and_save_filter(image, filter_name, filter_function):
    result = filter_function(image)
    output_path = f'{filter_name}_output.jpg'
    cv2.imwrite(output_path, result)
    print(f'{filter_name} filter applied and output saved to {output_path}')

# Apply and save different filters
apply_and_save_filter(img, 'original', lambda x: x)  # Original image

# Blur filters
apply_and_save_filter(img, 'gaussian_blur', lambda x: cv2.GaussianBlur(x, (5, 5), 0))
apply_and_save_filter(img, 'motion_blur', lambda x: cv2.filter2D(x, -1, np.ones((5, 5), np.float32) / 25))

# Sharpening filters
apply_and_save_filter(img, 'laplacian', lambda x: cv2.Laplacian(x, cv2.CV_64F))
apply_and_save_filter(img, 'unsharp_masking', lambda x: cv2.addWeighted(x, 1.5, cv2.GaussianBlur(x, (0, 0), 2), -0.5, 0))

# Edge detection filters
apply_and_save_filter(img, 'sobel_x', lambda x: cv2.Sobel(x, cv2.CV_64F, 1, 0, ksize=3))
apply_and_save_filter(img, 'prewitt_y', lambda x: cv2.filter2D(x, -1, np.array([[1, 1, 1], [0, 0, 0], [-1, -1, -1]])))

# Noise reduction filters
apply_and_save_filter(img, 'median_blur', lambda x: cv2.medianBlur(x, 5))
apply_and_save_filter(img, 'mean_blur', lambda x: cv2.blur(x, (5, 5)))

# Color filters
apply_and_save_filter(img, 'color_balance', lambda x: cv2.xphoto.createSimpleWB().balanceWhite(x))
apply_and_save_filter(img, 'histogram_equalization', lambda x: cv2.equalizeHist(cv2.cvtColor(x, cv2.COLOR_BGR2GRAY)))

# Morphological filters
kernel = np.ones((5, 5), np.uint8)
apply_and_save_filter(img, 'erosion', lambda x: cv2.erode(x, kernel, iterations=1))
apply_and_save_filter(img, 'dilation', lambda x: cv2.dilate(x, kernel, iterations=1))

# Display the original image
cv2.imshow('Original Image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()


import cv2
import numpy as np

# Read the input image
input_image_path = '/Users/nikhilvalsankulangareth/Downloads/original_image_1.jpg'
img = cv2.imread(input_image_path, cv2.IMREAD_COLOR)

# Ensure the image is loaded successfully
if img is None:
    print("Error: Unable to read the input image.")
    exit()

else:
    print("Img read")


# Define a function to apply and save a filter
def apply_and_save_filter(image, filter_name, filter_function):
    result = filter_function(image)
    output_path = f'/Users/nikhilvalsankulangareth/Downloads/{filter_name}_output.jpg'
    cv2.imwrite(output_path, result)
    print(f'{filter_name} filter applied and output saved to {output_path}')


# Apply and save different filters
apply_and_save_filter(img, 'original', lambda x: x)  # Original image

# Blur filters
apply_and_save_filter(img, 'gaussian_blur', lambda x: cv2.GaussianBlur(x, (5, 5), 0))
apply_and_save_filter(img, 'motion_blur', lambda x: cv2.filter2D(x, -1, np.ones((5, 5), np.float32) / 25))

# Sharpening filters
apply_and_save_filter(img, 'laplacian', lambda x: cv2.Laplacian(x, cv2.CV_64F))
apply_and_save_filter(img, 'unsharp_masking', lambda x: cv2.addWeighted(x, 1.5, cv2.GaussianBlur(x, (0, 0), 2), -0.5, 0))

# Edge detection filters
apply_and_save_filter(img, 'sobel_x', lambda x: cv2.Sobel(x, cv2.CV_64F, 1, 0, ksize=3))
apply_and_save_filter(img, 'prewitt_y', lambda x: cv2.filter2D(x, -1, np.array([[1, 1, 1], [0, 0, 0], [-1, -1, -1]])))



# Noise reduction filters
apply_and_save_filter(img, 'median_blur', lambda x: cv2.medianBlur(x, 5))
apply_and_save_filter(img, 'mean_blur', lambda x: cv2.blur(x, (5, 5)))

# Color filters
apply_and_save_filter(img, 'color_balance', lambda x: cv2.xphoto.createSimpleWB().balanceWhite(x))
apply_and_save_filter(img, 'histogram_equalization', lambda x: cv2.equalizeHist(cv2.cvtColor(x, cv2.COLOR_BGR2GRAY)))

# Morphological filters
kernel = np.ones((5, 5), np.uint8)
apply_and_save_filter(img, 'erosion', lambda x: cv2.erode(x, kernel, iterations=1))
apply_and_save_filter(img, 'dilation', lambda x: cv2.dilate(x, kernel, iterations=1))
```
